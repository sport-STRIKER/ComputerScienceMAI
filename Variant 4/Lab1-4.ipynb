7. Рассмотрим случайный вектор с формой 100х2, представляющей координаты, и найдем расстояния от точки к точке

````python
import numpy as np

def distance_matrix(points):
    if not isinstance(points, np.ndarray) or points.shape[1] != 2:
        print("Error: Input must be a NumPy array with shape (n_samples, 2).")
        return None
    
    n_samples = points.shape[0]
    distances = np.zeros((n_samples, n_samples))

    for i in range(n_samples):
        for j in range(n_samples):
            distances[i, j] = np.linalg.norm(points[i] - points[j])  

    return distances

np.random.seed(42)
points = np.random.rand(100, 2)

distance_matrix_result = distance_matrix(points)

if distance_matrix_result is not None:
    print("Example distances (top 5 rows):\n", distance_matrix_result[:5, :5])

    distances_point1 = distance_matrix_result[0]
    print(f"\nDistances for point 1:\n {distances_point1}")
````

9. Проанализируйте есть ли разница в цене между квартирами на 1 и 9 этажах

````python
import pandas as pd

df = pd.read_csv("russian-housing-market.csv")

first = df[df["floor"] == 1].index.tolist()
ninth = df[df["floor"] == 9].index.tolist()

first_price = 0
for i in range(len(first)):
    first_price += df.loc[first[i], "price_doc"]

ninth_price = 0    
for i in range(len(ninth)):    
    ninth_price += df.loc[ninth[i], "price_doc"]

if (ninth_price/len(ninth)) > (first_price/len(first)):
    print("Девятый дороже")

if (ninth_price/len(ninth)) < (first_price/len(first)):
    print("Первый дороже")

if (ninth_price/len(ninth)) == (first_price/len(first)):
    print("стоимости равны")
````

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinski Zaton) с течением времени

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("russian-housing-market.csv")

df["timestamp"] = pd.to_datetime(df["timestamp"])

df_nagatinskij = df[df["sub_area"] == "Nagatinskij Zaton"]

average_price_by_month = df_nagatinskij.groupby(pd.Grouper(key= "timestamp", freq = "M"))["price_doc"].mean()

average_price_by_month.plot(kind = "line")
plt.title("Средняя цена квартиры в Нагатинском Затоне по месяцам")
plt.xlabel("Дата")
plt.ylabel("Средняя цена")
plt.show()
````

2. Постройте гистограмму распределения цен на квартиры с общей площадью более 50 кв.метров

````python
import pandas as pd

df = pd.read_csv("russian-housing-market.csv")

df["timestamp"] = pd.to_datetime(df["timestamp"])

df_filtered = df[df["full_sq"] > 50]

plt.hist(df_filtered["price_doc"], bins = 5, edgecolor = "black")  

plt.xlabel("Цена")
plt.ylabel("Частота")
plt.title("Распределение цен на квартир с площадью более 50 кв.м.")
plt.show()
````

3. Визуализируйте количество проданных квартир в разных районах
